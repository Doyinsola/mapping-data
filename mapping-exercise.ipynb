```python
#importing pandas
import pandas as pd
import json
```

```python
#reading in the file as a dataframe
df = pd.read_csv("sample_data/wine-ratings.csv", index_col=0)
df.head()
```

```python
#Type of wine (red or white)
wine_type = df["variety"].unique()
print(wine_type)
```

```python
#Wine score range
print("The wine ratings are between",df['rating'].min(), "and", df['rating'].max())
```

```python
#The wines are from the following regions
wine_region = df['region'].unique()
print(wine_region)
```

```python
# Get all red and white wines
white_red_wines = df.loc[df['variety'].isin(['Red Wine','White Wine'])]
white_red_wines
```

```python
#Saved the filtered dataframe to a JSON file
white_red_wines.to_json('sample_data/white_red_wines.json', indent=2)
```
